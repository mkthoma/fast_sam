```python
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('fast_sam/examples/dog.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

original_h = image.shape[0]
original_w = image.shape[1]
print(original_w, original_h)
plt.figure(figsize=(10, 10))
plt.imshow(image)
```

```python
# Text prompt
!python FastSAM/Inference.py --model_path FastSAM.pt --img_path ./fast_sam/examples/dog.jpg  --text_prompt "the  dog"

image = cv2.imread('output/dog.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
original_h = image.shape[0]
original_w = image.shape[1]
print(original_w, original_h)
plt.figure(figsize=(10, 10))
plt.axis("off")
plt.imshow(image)
```
